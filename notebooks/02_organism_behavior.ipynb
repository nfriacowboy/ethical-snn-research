```python
# Create environment
env = GridWorld(size=(50, 50))

# Simulate behavior
for t in range(10):
    env_state = {
        'nearest_food': env.find_nearest_food(survival_org.position),
        'grid': env.grid
    }
    
    survival_org.step(env_state)
    print(f"Timestep {t}: Survival organism at {survival_org.position}, energy={survival_org.energy:.1f}")
```
